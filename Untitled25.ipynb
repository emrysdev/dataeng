```python
import pandas as pd

# Load the Excel file (replace with your file path)
file_path = 'C:/IT/Taf/merged_excel_data.xlsx'
excel_file = pd.ExcelFile(file_path)

# List all sheet names
print("Sheet names:", excel_file.sheet_names)

# Read all sheets into a list of DataFrames
dfs = [excel_file.parse(sheet_name) for sheet_name in excel_file.sheet_names]

# Outer join all DataFrames from different sheets
# Start by setting the first DataFrame as the base for joining
merged_df = dfs[0]

# Loop through all other DataFrames and perform outer join
for df in dfs[1:]:
    merged_df = pd.merge(merged_df, df, how='outer')

# Print the merged DataFrame after performing the outer join
print("\nMerged DataFrame after outer join:")
print(merged_df.head())

# Handle columns with similar names (e.g., 'income', 'income_1', etc.)
# Define a function to merge similar columns by summing them or using another aggregation function
def merge_similar_columns(df):
    # Iterate over columns and merge if they share the same prefix
    merged_columns = {}
    
    for col in df.columns:
        base_col_name = col.split('_')[0]  # Get the base name (before '_')
        if base_col_name not in merged_columns:
            merged_columns[base_col_name] = []
        merged_columns[base_col_name].append(col)
    
    # Combine similar columns by summing or using another method
    for base_col_name, cols_to_merge in merged_columns.items():
        if len(cols_to_merge) > 1:
            # Sum all columns with the same base name
            df[base_col_name] = df[cols_to_merge].sum(axis=1)
            # Drop the original columns that were merged
            df.drop(columns=cols_to_merge, inplace=True)
    
    return df

# Merge similar columns
merged_df = merge_similar_columns(merged_df)

# Print the final DataFrame after merging similar columns
print("\nDataFrame after merging similar columns:")
print(merged_df.head())

# Save the final merged DataFrame to a new Excel file
output_file_path = 'C:/IT/Taf/final_outer_join_merged_data.xlsx'
merged_df.to_excel(output_file_path, index=False)

print(f"Merged data saved to: {output_file_path}")
```
