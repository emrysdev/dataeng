```python
for num in range(0, 2001):
    if num % 5 == 0:
        print(num)
```

```python
import numpy as np

# Using list comprehension to calculate val/5 for each value in the range
result = [val/5 for val in np.arange(0, 2000)]

# Printing the output
print(result)
```
