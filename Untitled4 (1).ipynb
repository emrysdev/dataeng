```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
file = 'C:/Deakin/DataWrangling/week4/weather.csv'  # Update this path as needed
data = pd.read_csv(file)
data
```

```python
# Step 2: Convert Columns to Metric Units
# Convert temperature and dew point to Celsius, precipitation to millimeters, visibility to meters, 
# and wind speed and gust to meters per second.
# Conversion functions
def f_to_c(f):
    return (f - 32) * 5.0/9.0

def inches_to_mm(inches):
    return inches * 25.4

def miles_to_meters(miles):
    return miles * 1609.34

def mph_to_mps(mph):
    return mph * 0.44704

# Apply conversions
data['temp'] = data['temp'].apply(f_to_c)
data['dewp'] = data['dewp'].apply(f_to_c)
data['precip'] = data['precip'].apply(inches_to_mm)
data['visib'] = data['visib'].apply(miles_to_meters)
data['wind_speed'] = data['wind_speed'].apply(mph_to_mps)
data['wind_gust'] = data['wind_gust'].apply(mph_to_mps)

# Verify the conversions
data.head()
```

```python
# Step 3: Compute Daily Mean Wind Speeds for LGA
# Group the data by year, month, and day, then compute the daily mean wind speeds for LGA.

# Filter data for LGA
lga_data = data[data['origin'] == 'LGA']

# Compute daily mean wind speeds
lga_daily_mean_wind_speed = lga_data.groupby(['year', 'month', 'day'])['wind_speed'].mean().reset_index()

# Convert to a more readable date format
lga_daily_mean_wind_speed['date'] = pd.to_datetime(lga_daily_mean_wind_speed[['year', 'month', 'day']])
lga_daily_mean_wind_speed = lga_daily_mean_wind_speed[['date', 'wind_speed']]

# Display the results
lga_daily_mean_wind_speed.head()
```

```python
# Step 4: Plot Daily Mean Wind Speeds
# Create a plot to visualize the daily mean wind speeds at LGA.

# Plot daily mean wind speeds
plt.figure(figsize=(14, 7))
plt.plot(lga_daily_mean_wind_speed['date'], lga_daily_mean_wind_speed['wind_speed'], label='Daily Mean Wind Speed')
plt.xlabel('Date')
plt.ylabel('Wind Speed (m/s)')
plt.title('Daily Average Wind Speed at LGA in 2013')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Step 5: Identify the Ten Windiest Days
# Identify the ten windiest days at LGA based on the mean daily wind speeds.

# Identify the ten windiest days
windiest_days = lga_daily_mean_wind_speed.nlargest(10, 'wind_speed')

# Display the ten windiest days
print(windiest_days)
```

```python
# Q5: Compute monthly mean wind speeds for all three airports, locate and replace outliers
def detect_and_replace_outlier(data):
    Q1 = data['wind_speed'].quantile(0.25)
    Q3 = data['wind_speed'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data['wind_speed'] < lower_bound) | (data['wind_speed'] > upper_bound)]
    data.loc[(data['wind_speed'] < lower_bound) | (data['wind_speed'] > upper_bound), 'wind_speed'] = np.nan
    return data, outliers

# Detect and replace outliers for each airport
jfk_data, jfk_outliers = detect_and_replace_outlier(data[data['origin'] == 'JFK'])
ewr_data, ewr_outliers = detect_and_replace_outlier(data[data['origin'] == 'EWR'])
lga_data, lga_outliers = detect_and_replace_outlier(data[data['origin'] == 'LGA'])

# Compute monthly mean wind speeds
monthly_mean_wind_speed = data.groupby(['year', 'month', 'origin'])['wind_speed'].mean().reset_index()
monthly_mean_wind_speed['date'] = pd.to_datetime(monthly_mean_wind_speed[['year', 'month']].assign(day=1))
monthly_mean_wind_speed = monthly_mean_wind_speed.pivot(index='date', columns='origin', values='wind_speed')
monthly_mean_wind_speed

```

```python
# Q6: Plot monthly mean wind speeds for all three airports
plt.figure(figsize=(14, 7))
plt.plot(monthly_mean_wind_speed.index, monthly_mean_wind_speed['JFK'], label='JFK')
plt.plot(monthly_mean_wind_speed.index, monthly_mean_wind_speed['EWR'], label='EWR')
plt.plot(monthly_mean_wind_speed.index, monthly_mean_wind_speed['LGA'], label='LGA')
plt.xlabel('Month')
plt.ylabel('Wind Speed (m/s)')
plt.title('Monthly Average Wind Speed in 2013')
plt.legend()
plt.grid(True)
plt.show()

# Optional Features (For postgraduate students)
# 1. List all missing temperature readings for JFK
jfk_missing_temps = jfk_data[jfk_data['temp'].isna()]
print("Missing temperature readings for JFK:")
print(jfk_missing_temps)

# 2. Add missing records for JFK
full_date_range = pd.date_range(start='2013-01-01', end='2013-12-31', freq='H')
jfk_data_full = jfk_data.set_index('time_hour').reindex(full_date_range).reset_index().rename(columns={'index': 'time_hour'})
jfk_data_full['origin'] = 'JFK'
jfk_data_full[['year', 'month', 'day', 'hour']] = jfk_data_full['time_hour'].apply(lambda x: [x.year, x.month, x.day, x.hour]).apply(pd.Series)

# 3. Compute daily average temperatures with linear interpolation
jfk_data_full['temp'] = jfk_data_full['temp'].interpolate()

# Compute daily average temperatures
jfk_daily_avg_temp = jfk_data_full.groupby(jfk_data_full['time_hour'].dt.date)['temp'].mean().reset_index(name='temp')
jfk_daily_avg_temp['time_hour'] = pd.to_datetime(jfk_daily_avg_temp['time_hour'])

# 4. Plot comparison of missing value-omitted vs linearly interpolated temperatures
plt.figure(figsize=(14, 7))
plt.plot(jfk_daily_avg_temp['time_hour'], jfk_daily_avg_temp['temp'], label='Interpolated Temp')
plt.xlabel('Date')
plt.ylabel('Temperature (Â°C)')
plt.title('Daily Average Temperatures at JFK in 2013 (Interpolated)')
plt.legend()
plt.grid(True)
plt.show()
```
